```python
import os
import xarray as xr
import glob

load_path = 'C:/Users/GUIDO/Master/MSWEP'
save_path = 'data/MSWEP_aggregated_3h'
filenames = glob.glob(os.path.join(load_path, "*.nc"))

concatenated = xr.open_dataset(filenames[0])
past_month = str(concatenated.time.data[0])[5:7]
past_year = str(concatenated.time.data[0])[:4]

for filename in filenames[1:]:
    with xr.open_dataset(filename) as new_df:
        current_month = str(new_df.time.data[0])[5:7]
        current_year = str(new_df.time.data[0])[:4]
        
        if current_month == past_month and current_year == past_year:
            concatenated = xr.concat([concatenated, new_df], dim='time')
        else:
            concatenated.to_netcdf(f'{save_path}/mswep_{past_year}_{past_month}.nc')
            concatenated = new_df
            print(f'DONE: {past_year}_{past_month}')
        past_month = current_month
        past_year = current_year 
# save last one, which doesn't get saved because the if condition under which is the save doesn't apply to the last month
concatenated.to_netcdf(f'{save_path}/mswep_{past_year}_{past_month}.nc')
```
